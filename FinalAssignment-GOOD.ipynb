```python
"""Creates pandas from scratch, taken from Week 8 lecture"""
import csv

def tryAsNumber(string):
    try:
        return int(string)
    except:
        try:
            return float(string)
        except:
            return string
        
def loadCSVDataFile(filenameOrURL):
    data = []   
    with open(filenameOrURL) as csvfile:
        datareader = csv.reader(csvfile)
        firstline = True
        for record in datareader:
            if firstline:
                firstline = False
                continue
            if len(record[:3])>0:
                data.append([tryAsNumber(val) for val in record[:3]])
    return data

```

```python
"""Remove movies allready watched from list of all movies. 
    takes input userX and movieRatingsData
    outputs set of unwatched moviesID for userX"""

def removeAllReadyWatched(userX,movieRatingsData):
    mySetMovies = getUserXMovies(userX,movieRatingsData)
    usersIDs,moviesIDs = getSetIDs(movieRatingsData)
    for myMovie in mySetMovies:
        moviesIDs.remove(myMovie)
    return len(moviesIDs)


unwatchedList = removeAllReadyWatched(1,movieRatingsData)
print(unwatchedList)
#print(movieRatingsData)
#print(movieRatingsData[1])
#print(movieRatingsData[1][1])
```

```python
def retrieveVector(movieID,matrix):
    """retrieve a vector for a given movieID from the matrix """
    for movie in matrix:
        if (movieID == movie[0]):
            return movie
        
retrieveVector(2,matrix)       
```

```python
def calculateDistance(vector1, vector2):
    
    """A function that calculates the distance between two vectors 
    (movie rating 1 and movie rating 2) this is the first step before 
    creating the function that find distance between Top Rated Movie and 
    all other movies. vector1 and vector2 are simply movie ratings in this case."""
    
    distance=0 #You compare two vectors so default value you begin at 0
    for v in range(len(vector1)):
        #print( vector1[v],vector2[v]) #to make sure the all element is being calculated
        distance += ((vector1[v]-vector2[v])**2)    
    return distance ** 0.5
    
    
calculateDistance([-7,-10],[-8,7])    
```

```python
"""This is the second step where I will find the distance between the Top Rated Movie for user x and
all movies in the data set, besides the ones user x has already seen. We will use the calculateDistance function defined before"""

def allDistances(myVector,dataMatrix,myMovieSet):
    
    distance = [] 
    """Creates a list where all distances between myVector and the data in Movie Lens (dataMatrix) 
    will be recorded. This will be recorded in the same order as the records in the Movie lens data set"""
    
    for Index in range(len(dataMatrix)):
        if not (dataMatrix[Index][0] in myMovieSet):
            distance.append((dataMatrix[Index][0],calculateDistance(myVector[1:], dataMatrix[Index][1:])))
    return distance[1:]

distances = allDistances(matrix[1],matrix,[1,2,3,4,5])
```

```python
"""Returns top k recommended movies for userX"""

def getTopRecomendations(userX,k):
    
    # initializes empty recommendations list(to fill up with recomendations)
    recommendations = []
    
    # Read Data
    movieIDData=loadCSVDataFile('./data/movies.csv')
    movieRatingsData= loadCSVDataFile('./data/Ratings100USERS.csv')
    
    # generate matrix
    matrix = genMovieMatrix(movieRatingsData)
    
    # get UserX's top rated movie
    movieID = getTopMovieUserX(userX,movieRatingsData)
    #print("Top movie for userX has ID: ", movieID)
    
    # Retrieve movie vector based on movieID(Selected by getTopMovieUserX function{esun riba esaki})
    myMovieVector = retrieveVector(movieID,matrix)
    #print("Top movie vector: ", myMovieVector)
    
    # Retrieve all movies watched by userX (This is necesarry inorder to not recommend a allready watched movie)
    userXMovies = getUserXMovies(userX,movieRatingsData)
    #print("Movies all ready watched by userX(MovieID): ", userXMovies)
    
    # Calculates the distance between all movies and userX's top movie(Skips movies allready watched)
    distances = allDistances(myMovieVector,matrix,userXMovies)
    
    # Sorts the most similar movies to my top movie
    for recommendation in getNearestNeighbour(distances,k):   
        recommendations.append(recommendation[0])
        
    # changes recommendations list from movieID to movieTitle     
    recommendations = getMovieName(recommendations,movieIDData)
    return recommendations

getTopRecomendations(110,20)
```
